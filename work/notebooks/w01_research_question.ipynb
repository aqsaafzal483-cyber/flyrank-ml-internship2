## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

**Decision:**
The project helps the content management team decide which pages should be refreshed first.

**Action:**
Editors and SEO specialists will update the pages with the highest opportunity scores by improving content and optimizing keywords.

**Cost of a wrong recommendation:**
If the model recommends refreshing a page that does not need updating, time and resources are wasted. If the model misses a page that needs refreshing, the page may remain outdated and lose traffic or engagement.

**Why machine learning can help:**
Large websites contain many pages, making manual review difficult. Machine learning can analyze historical data and help prioritize the pages most likely to benefit from a refresh.
